```python
# 步骤1: 按用户和日期分组统计购买次数
df_gap = (
    df_paid.groupby(['用户名', 'date'])['商品名']  # 按用户+日期分组
    .count()                                       # 统计购买次数
    .reset_index()                                 # 恢复为DataFrame
    .sort_values(by=['用户名', 'date'])            # 按用户和时间排序
)

# 步骤2: 计算每个用户每天购买次数的差值（与前一天对比）
df_gap['gap'] = (
    df_gap.groupby('用户名')['商品名']  # 按用户分组
    .diff()                             # 计算差值（后一天 - 前一天）
    .fillna(0)                          # 填充NaN为0（第一个日期无差值）
)
df_gap
```

```python
# 步骤1：统计各环节的唯一用户数
df0 = df['用户名'].nunique()                     # 浏览用户数（行为类型不筛选，假设为总用户）
df1 = df[df.行为类型 == 1]['用户名'].nunique()    # 点击用户数（行为类型=1）
df2 = df[df.行为类型 == 2]['用户名'].nunique()    # 加购用户数（行为类型=2）
df3 = df[df.行为类型 == 3]['用户名'].nunique()    # 收藏用户数（行为类型=3）
df4 = df[df.行为类型 == 4]['用户名'].nunique()    # 购买用户数（行为类型=4）

# 步骤2：计算并打印流失率（保留两位小数）
print(f'浏览-点击的消失率：{( (df0 - df1) / df0 * 100 ):.2f}%')
print(f'点击-加购的流失率：{( (df1 - df2) / df1 * 100 ):.2f}%')
print(f'加购-收藏的流失率：{( (df2 - df3) / df2 * 100 ):.2f}%')
print(f'收藏-购买的流失率：{( (df3 - df4) / df3 * 100 ):.2f}%')
```

```python
import datetime
df_recency=df_paid.groupby('用户名')['date'].max()
df_recency=(pd.to_datetime(datetime.date.today())-df_recency).apply(lambda x:x.days)
df_recency_avg=df_recency.mean()
df_recency[df_recency<df_recency_avg]=0
df_recency[df_recency>=df_recency_avg]=1
df_recency
```

```python
df_paid.head()
```

```python
df_frequency=df_paid.groupby('用户名')['商品名'].count()
df_frequency[df_frequency<df_frequency.mean()]=0
df_recency[df_frequency>=df_frequency.mean()]=1
df_frequency
```

```python
# 假设 df_recency 和 df_frequency 已经定义
df_rf = pd.concat([df_recency, df_frequency], axis=1)

# 重命名列名
df_rf.columns = ['recency', 'frequency']

# 将 'recency' 和 'frequency' 列转换为字符串类型
df_rf['recency'] = df_rf['recency'].astype(str)
df_rf['frequency'] = df_rf['frequency'].astype(str)

# 合并 'recency' 和 'frequency' 列，生成新列 'RF'
df_rf['RF'] = df_rf['recency'] + df_rf['frequency']

# 显示最终的 DataFrame
df_rf
```

```python
def trans_label(x):
    if x=='00':
        return '重要挽留客户'
    elif x=='01':
        return '重要保持客户'
    elif x=='02':
        return '重要发展客户'
    else:
        return '重要价值客户'
```

```python
df_rf['label']=df_rf['RF'].apply(trans_label)
df_rf
```

```python
# 按照商品种类分组并计算购买量
df_buy = df[df.行为类型==4].groupby('商品种类')['用户名'].count().reset_index()

# 修改列名
df_buy.columns = ['商品种类', '购买量']

# 根据购买量排序
df_buy.sort_values(by='购买量', ascending=False, inplace=True)

# 计算累计购买量
df_buy['累计购买量'] = df_buy['购买量'].cumsum()

# 计算购买量总和的 80% 的值
value = 0.8 * df_buy['购买量'].sum()

# 根据累计购买量给商品打标签
df_buy['label'] = df_buy['累计购买量'].apply(lambda x: '前80%' if x <= value else '后20%')

df_buy
```
